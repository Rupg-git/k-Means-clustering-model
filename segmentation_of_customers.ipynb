```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/Mall_Customers.csv')
```

```python
from pandas.core.resample import date_range
X=df.iloc[:,[3,4]].values
```

```python
print(X)
```

```python
wcss=[]
```

```python
for i in range(1,11):
  kmeans = KMeans(n_clusters=i , init='k-means++',random_state = 42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('the elbow graph')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=0)
Y=kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=60,c='green',label='cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=60,c='blue',label='cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=60,c='red',label='cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=60,c='orange',label='cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=60,c='purple',label='cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,c='yellow',label='centroids')
plt.title('customer groups')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.show()
```
